```python
import numpy as np
import scipy.io.wavfile
import matplotlib.pyplot as mp
import scipy.signal as signal

# Synthesize a square wave at 440Hz for a duration of 2 seconds.

fs = 44100     # in samples/second
duration = 8   # in seconds
frequency = 220

t = np.linspace(0, duration, int(duration*fs))

#mysquarewave = signal.square(2*np.pi*t*frequency)
#myantialias = signal.decimate(mysquarewave, 5, 8, 'iir')

mywave = (np.sin(2*np.pi*t*frequency)+1/3*np.sin(2*np.pi*3*t*frequency)+1/5*np.sin(2*np.pi*5*t*frequency)+1/7*np.sin(2*np.pi*7*t*frequency)+1/9*np.sin(2*np.pi*9*t*frequency))

# For plotting, we don't need to see so many samples
stopAtThisSample = 1000
mp.plot(t[1:stopAtThisSample], mywave[1:stopAtThisSample])
mp.xlabel('Time [sec]')

# Turn down the amplitude somewhat so it's not too loud
gain = 1
mywave = mywave * gain

intSignal = np.int16(mywave * 32767)
scipy.io.wavfile.write('squarealiasGibbs.wav', fs, intSignal)
```
